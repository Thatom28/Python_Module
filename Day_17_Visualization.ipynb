## Bar graph

```python
tips_df = pd.read_csv('tips.csv')
tips_df
```

```python
tips_df['tip_percentage'] = ((tips_df['tip'] / tips_df['total_bill']) * 100).round(2)
# tips_df['tip_percentage'] = ((tips_df['tip'] / tips_df['total_bill']) * 100,2)  #sam as above
tips_df
```

```python
# x_days = tips_df['day'].unique()  #what you want for teh x-axis
y_avg_total_bill = tips_df.groupby('day')['total_bill'].mean()  # what to plot for the y-axis
x_days = y_avg_total_bill.index

plt.bar(x_days, y_avg_total_bill, color='crimson', alpha= 0.5)   #Draws the graph, you can change the color
plt.xlabel("Days")  # label the x-axis
plt.ylabel("average sales in R")  #labe the y axis
plt.title('Average sales in the week')
```

## pie charts
- to get the percentage autopct='%1.1f%%'

```python
y_avg_total_bill.plot(kind = 'pie',autopct='%2.1f%%', xlabel ="Days", ylabel="Average sales in R", title="Average sales in a week", color='crimson')
```

## scatter plot: why it is used
- used to show the outliers
- and shows the treands

```python
#Tip percentage vs total_bill
plt.scatter(tips_df['total_bill'], tips_df['tip_percentage'])
plt.xlabel("Tip percentage %")  # label the x-axis
plt.ylabel("Total bill (R)")  #labe the y axis
plt.xticks(rotation=45) #rotate the labels
```

## Histogram -> Ticket_size
how big is your order size
how much the order is
ticket-size - total_-bill -> x.axis vs frequency -> y.axis

```python
plt.hist(tips_df['total_bill'], color ='orange', bins=50)  #bins -> how much to divide
plt.xlabel('Total bill (R)')
plt.ylabel('Frequency')
```

## seaborn
- the stick represent the minimum and th maximum values and the bar is the average

```python
# x_days vs y_total_bill
# automatically calculates the average by the x-axis
# the lines represent the minimum and the maximum of the days
sns.barplot(x ='day', y ='total_bill', data=tips_df, palette='pastel')
plt.xlabel('Average sales (R)')
plt.ylabel('days')
plt.title("")
```

```python
sns.violinplot(x = 'total_bill', y ='tip', data= tips_df,color='green')
plt.xlabel('Tip (R)')
plt.ylabel('Total bill (R)')
plt.title("Total bill vs. Tip with Regression Line")
```
